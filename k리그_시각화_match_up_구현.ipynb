```python
import pandas as pd
import json

csv_path = "/content/team_vs_team_advanced.csv"
df = pd.read_csv(csv_path)

print("컬럼 목록:")
print(df.columns.tolist())

# ✅ 네 CSV에 맞춘 컬럼 매핑
COL_TEAM = "team_name_ko"
COL_OPP  = "opp_team_name_ko"

COL_DTSS = "dTSS_mean"
COL_DSGP = "dSGP_mean"
COL_DPTI = "dPTI_mean"

COL_N    = "n"
COL_PPG  = "pts_per_game"
COL_WIN  = "win_rate"

use_cols = [COL_TEAM, COL_OPP, COL_DTSS, COL_DSGP, COL_DPTI, COL_N, COL_PPG, COL_WIN]

df_use = (
    df[use_cols]
    .dropna(subset=[COL_TEAM, COL_OPP])
    .fillna(0)
)

records = []
for _, r in df_use.iterrows():
    records.append({
        "team": str(r[COL_TEAM]),
        "opp": str(r[COL_OPP]),
        "dTSS_mean": float(r[COL_DTSS]),
        "dSGP_mean": float(r[COL_DSGP]),
        "dPTI_mean": float(r[COL_DPTI]),
        "n": int(r[COL_N]),
        "ppg": float(r[COL_PPG]),
        "win_rate": float(r[COL_WIN]),
    })

out_path = "/content/team_vs_team_advanced.json"
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(records, f, ensure_ascii=False, indent=2)

print(f"✅ JSON 생성 완료: {out_path}")
print(f"총 매치업 수: {len(records)}")
print("샘플 3개:")
print(records[:3])
```
